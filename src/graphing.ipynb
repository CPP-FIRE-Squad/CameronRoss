# Bonus - Graphing m10q_res250

```python
import matplotlib.pyplot as plt
from stellarutil import Simulation
from IPython.display import clear_output
import joblib, numpy as np

def graph(x,y,z,c, title, Y, Y_pred):
    # Create a figure
    fig = plt.figure(figsize=(12, 10))
    plt.suptitle(title, fontsize=16)
    
    # 2D Scatter Plot
    plt.subplot(2, 2, 1)
    plt.title("2D Star Positions")
    plt.scatter(x, y, label='Data Points', c=c, vmin=0, vmax=100)
    plt.colorbar()
    plt.xlabel('X')
    plt.ylabel('Y')

    # 3D Sub Plot
    ax = plt.subplot(2, 2, 2, projection='3d')    
    sc = ax.scatter(x, y, z, c=c, marker='o',vmin=0, vmax=100)
    plt.title("3D Star Positions")
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')
    fig.colorbar(sc)

    # Predicted vs Actual
    plt.subplot(2,2,3)
    plt.title("Predicted vs Actual Dark Matter Masses")
    plt.scatter(Y, Y_pred, label='Data Points', c=c, vmin=0, vmax=100)
    plt.colorbar()
    # Get the max and min value and plot y=x line
    minVal = min(min(Y), min(Y_pred))
    maxVal = max(max(Y), max(Y_pred))
    plt.plot([minVal, maxVal], [minVal, maxVal], color='green', label='y = x')
    # Add labels and legend
    plt.xlabel('Actual Mass [M☉]')
    plt.ylabel('Predicted Mass [M☉]')
    plt.legend()
    plt.loglog()

    # Bin
    plt.subplot(2,2,4).hist(c, bins=20)
    plt.title("Histogram of Percent Differences")
    
    # Show the plots
    plt.tight_layout()
    txt=f"Halo 0 has {len(Y)} stars. The average percent difference is {np.mean(c):.2f}%."
    plt.figtext(0.5, -0.02, txt, wrap=True, horizontalalignment='center', fontsize=12)
    plt.show()
    fig.savefig(f"../images/{title}_plot.png", bbox_inches="tight")
```

```python
def graphSim(simName, nickName):
    # Load the sim
    sim = Simulation(simulation_name=simName, species=['star'])
    halo = sim.get_halo()
    # Load the data
    a = nickName.split("_")[0][0].upper() + nickName.split("_")[1][0].upper()
    X = joblib.load(f'../data/pickle/{nickName}/X_{a}_0_to_100.pkl')[:len(halo.stars)] # 15649 is the number of stars at halo 0
    Y = joblib.load(f'../data/pickle/{nickName}/Y_{a}_0_to_100.pkl')[:len(halo.stars)] # 15649 is the number of stars at halo 0
    # Get predictions
    knn_regressor = joblib.load("model_sigma.pkl")
    Y_pred = np.array(knn_regressor.predict(X))
    # Get the x, y, and z coordinates of the stars
    x = [star.x for star in halo.stars]
    y = [star.y for star in halo.stars]
    z = [star.z for star in halo.stars]
    c = (Y / Y_pred)
    c = (np.abs(Y - Y_pred) / ((Y + Y_pred) / 2)) * 100
    graph(x, y, z, c, simName,Y, Y_pred)

graphSim("m10q_res250md", "little_quebec")
graphSim("m10r_res250md", "little_romeo")
graphSim("m10v_res250md", "little_victor")
graphSim("m10v_res030md", "big_victor")

```
